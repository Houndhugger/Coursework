```python
import numpy as np
import xlrd
import pandas as pd
book = xlrd.open_workbook('kinetic data complete.xlsx')
sheet = book.sheet_by_name('Sheet1')
data = [[sheet.cell_value(r, c) for c in range(0,3)] for r in range(0,312)]
data2 = [[sheet.cell_value(r, c) for c in range(3,4)] for r in range(0,312)]
test_data = np.array(data)
light_data = np.array(data2)
```

```python
writer1 = pd.ExcelWriter('generateddatakinetic.xlsx', engine='xlsxwriter')
for i in range(1, 26):
    b = np.random.normal(test_data,test_data*0.05*0.333,(312,3))
   
    df = pd.DataFrame(b)
    df2 = pd.DataFrame(light_data)
    startrow = (i-1)*312
    df.to_excel(writer1,startrow=startrow, header=False, index=False)
    df2.to_excel(writer1,startrow=startrow,startcol=3, header=False, index=False)
    c = np.random.normal(test_data,test_data*0.03*0.333,(312,3))
    df = pd.DataFrame(c)
    startrow = (i-1)*312+25*312
    df.to_excel(writer1,startrow=startrow, header=False, index=False)
    df2.to_excel(writer1,startrow=startrow,startcol=3, header=False, index=False)
writer1.save()
```

```python
book = xlrd.open_workbook('generateddatakineticwithgradient.xlsx')
sheet = book.sheet_by_name('Sheet1')
data3 = [[sheet.cell_value(r, c) for c in range(0,7)] for r in range(0,15600)]
complete_data = np.array(data3)
prepared_data = np.delete(complete_data, list(range(25, complete_data.shape[0], 26)), axis=0)
writer2 = pd.ExcelWriter('prepareddatakinetic.xlsx', engine='xlsxwriter')
df3 = pd.DataFrame(prepared_data)
df3.to_excel(writer2, header=False,index=False)
writer2.save()
```
